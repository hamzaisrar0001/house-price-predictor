```python
labels = data['price']
# conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date'] = pd.to_datetime(data['date'])
data['date'] = (data['date'].dt.year == 2014).astype(int)
# data['date'] = conv_dates
train1 = data.drop(['id', 'price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2,
          learning_rate = 0.1,loss='squared_error')
```

```python
clf.fit(x_train, y_train)
```

```python
# clf = ensemble.GradientBoostingRegressor(
#     n_estimators=400,
#     max_depth=5,
#     min_samples_split=2,
#     learning_rate=0.1,
#     loss='squared_error'
# )

# clf.fit(x_train, y_train)

# t_sc = np.zeros(clf.n_estimators, dtype=np.float64)
# testsc = np.arange(clf.n_estimators) + 1
clf = ensemble.GradientBoostingRegressor(
    n_estimators=400,
    max_depth=5,
    min_samples_split=2,
    learning_rate=0.1,
    loss='squared_error'
)

clf.fit(x_train, y_train)

# t_sc ab sahi tarike se fill hoga
from sklearn.metrics import r2_score

t_sc = np.array([r2_score(y_test, y_pred) 
                 for y_pred in clf.staged_predict(x_test)])

train_sc = np.array([r2_score(y_train, y_pred) 
                     for y_pred in clf.staged_predict(x_train)])

testsc = np.arange(clf.n_estimators) + 1
```

```python
# plt.figure(figsize=(12, 6))
# plt.subplot(1, 2, 1)
# plt.plot(testsc,clf.train_score_,'b-',label= 'Set dev train')
# plt.plot(testsc,t_sc,'r-',label = 'set dev test') 
plt.figure(figsize=(12, 6))
plt.plot(testsc, train_sc, 'b-', label='Training Score')
plt.plot(testsc, t_sc,     'r-', label='Test Score')
plt.xlabel('Number of Estimators')
plt.ylabel('R² Score')
plt.title('Gradient Boosting — Train vs Test Score')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA

pca = PCA()
pca.fit_transform(scale(train1))

# Kitna variance har component explain karta hai
explained_variance = pca.explained_variance_ratio_

# Plot
plt.figure(figsize=(10, 5))
plt.bar(range(1, len(explained_variance)+1), 
        explained_variance, alpha=0.7, color='steelblue')
plt.plot(range(1, len(explained_variance)+1), 
         np.cumsum(explained_variance), 'ro-', label='Cumulative')
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.title('PCA — Variance Explained by Each Component')
plt.legend()
plt.axhline(y=0.95, color='green', linestyle='--', label='95% threshold')
plt.show()

# Print karo kitne components 95% variance cover karte hain
n_components_95 = np.argmax(np.cumsum(explained_variance) >= 0.95) + 1
print(f"95%  for variance {n_components_95} components is enough")
print(f"Total features: {train1.shape[1]}")
```

## PCA Insight
Out of 19 features, 14 principal components capture 95% of the variance.
This suggests moderate multicollinearity among features like 
sqft_living, sqft_above, and sqft_basement.
